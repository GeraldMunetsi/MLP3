```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10, 20)
        self.fc2 = nn.Linear(20, 5)

    def forward(self, x):
        x = F.relu(self.fc1(x))  # [B,10] → [B,20]
        x = self.fc2(x)          # [B,20] → [B,5]
        return x                 # now matches y.shape
```

```python
#Training the model
#Internally PyTorch tracks:

model=SimpleNN()

x=torch.randn(1,10)
model.parameters()
print(model)

optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3)
criterion=nn.MSELoss()

x=torch.randn(10,10)
y=torch.rand(10,5)

optimizer.zero_grad()
pred=model(x)

loss =criterion(pred,y)
loss.backward()  #It computes gradients using automatic differentiation. Each parameter has a gradient stored.
optimizer.step() #UDATES PARAMETERS



```

```python
import torch
import torch.nn as nn

#You just built a linear regression neural network layer.
class LinearModel(nn.Module):

    def __init__(self,input_size,output_size):
        super().__init__() # takes x and input w*x+b

        #nn.Parameter wraps our normal tesors and tells pytorch that they are our nn.Modules parameters to be optimizwed

        self.w = nn.Parameter(torch.randn(output_size,input_size))
        self.b = nn.Parameter(torch.zeros(output_size))

    def forward(self, x):
        return x@self.w.T + self.b #always start by multiplying x
    

input_data=torch.randn(10,5)

#creating an instance of our model
linear_model= LinearModel(5,1) # 5 features, output size =1, my model predicts one number per input sample , there is no batch 

#So 10 rows = 10 inputs processed together, bactch processing, recall: (batch_size , features)
#perform the forward pass
output=linear_model(input_data)

print(output.shape)
print(output.detach())# detarch disconnects the tensor and returns it with no history of previous calculationa


#Output tensor.Size([10,1])


```

```python
#Pytorch for NN
#nn.Linear(no of inputs, number of outputs)
import torch
import torch.nn as nn
linear=nn.Linear(3,1) # pytorch initialize aoutomatically weights using Kaiming Uniform Initialization

#W,b∼U(−srt(k),sqrt(k)) , k=1/3, weights are sampled from W∼U(−sqr(1/5)​,sqr(1/5​)

#to avaoid eploding and vanishing of gradients
# So initialization keeps the variance of signals stable across layers.
# This idea comes from He initialization (2015).

print(linear.weight.data)
print(linear.bias.data)
#pytorch initializes grad to note not zero, they get created after the first backwards pass
print(linear.weight.grad)
print(linear.bias.grad)

```

```python
#create a random data input tensor

data=torch.randn(100,3)

#create some noisey target data
target=data.sum(1,keepdims=True)+ 0.01*torch.randn(data.shape[0],1)

print(data[:10])# input data 
print(target[:10])

# to perform forward pass we just need to call the function

target_pred =linear(data)
print(target_pred,data[:10])







```

```python
#Plotting , inputs vs targets
import matplotlib.pyplot as plt
#uSE the outputs
plt.scatter(data[:,0],target_pred.detach())

#Use the ground thruth
plt.scatter(data[:,0],target,marker="x")
plt.legend("Predictions","Ground Thruth")
plt.xlabel("Inputs")
plt.ylabel("Outputs")
```

```python
#Loss functions
loss_function=nn.MSELoss()

optimizer=torch.optim.SGD(linear.parameters(),lr=0.01) # we are telling the GD whta it needs to optimize
#so we pass it are the linear parameters

loss=loss_function(target_pred,target)

print(loss.item) #averaged loss for the entire bactch
loss.backward()

print(linear.weight.grad)


#tell the optimizer to perform an update step
optimizer.step()

#Perform another forward pass to the model to check the the new loss
target_pred=linear(data)
loss=loss_function(target_pred,target)
print(loss.item())
```

```python
#We now need a Trining loop untill the loss is low and the model cant learn more

loss_logger=[]

for i in range(1000):
    target_pred=linear(data)
    loss=loss_function(target_pred,target)
    optimizer.zero_grad() # sets the stored gradients to zero
    loss.backward()
    optimizer.step()

    loss_logger.append(loss.item())

print("Loss",loss.item())
```
